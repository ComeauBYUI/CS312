<h1>SYLLABUS</h1>
<h2>Course Overview</h2>
<p>CS 124 is the first class in a <g class="gr_ gr_141 gr-alert gr_spell gr_inline_cards gr_run_anim ContextualSpelling ins-del multiReplace" id="141" data-gr-id="141">two class</g> sequence exploring how to write code in C++. The first class, CS 124, teaches us how to write procedural programs. The second class, CS 165, teaches us how to write <g class="gr_ gr_142 gr-alert gr_spell gr_inline_cards gr_run_anim ContextualSpelling ins-del multiReplace" id="142" data-gr-id="142">object oriented</g> programs. The goal of CS 124 is that each student will be able to solve problems in C++ and have a solid foundation in software development methodology. By the end of the semester, each student will:</p>
<ul>
<li>Be well prepared for CS 165 and the other courses in the CS major</li>
<li>Have confidence in their ability to solve problems with C++</li>
<li>Possess a tool-bag of different approaches to <g class="gr_ gr_175 gr-alert gr_gramm gr_inline_cards gr_run_anim Grammar multiReplace" id="175" data-gr-id="175">solve</g> software problems</li>
<li>Have a firm foundation in the basic constructs of procedural C++</li>
</ul>
<p>These goals will be explored in the context of C++ using the Linux operating system.</p>
<h2>Course Outline</h2>
<p>This class will be divided into four units, each of which <g class="gr_ gr_134 gr-alert gr_gramm gr_inline_cards gr_run_anim Grammar multiReplace" id="134" data-gr-id="134">working</g> the same way. After all the activities in a section are completed, you will have the opportunity to take the section test and complete the project. The four units are:</p>
<ul>
<li><b>Simple Programs.</b> How to get a C++ program to run on a Linux system and basic input/output. This includes branching mechanisms, functions, and expressions.</li>
<li><b>Design &amp; Loops.</b> The goal of this section is to be able to solve problems using loops and to learn how to tackle large and complex problems.</li>
<li><b>Pointers &amp; Arrays.</b> <g class="gr_ gr_131 gr-alert gr_gramm gr_inline_cards gr_run_anim Punctuation only-ins replaceWithoutSep" id="131" data-gr-id="131">Next</g> we will learn how to handle and manipulate large amounts of data in increasingly complex data structures. Specifically, we will learn about arrays and file I/O.</li>
<li><b>Advanced Topics.</b> The final section is designed to reinforce the core skills learned in earlier sections.</li>
</ul>
<h2>Class Time</h2>
<p>Each unit will be sub-divided into seven or eight chapters / class-periods. We will work through one chapter per class period. Each class period (with certain exceptions) will follow the following pattern:</p>
<ol>
<li><b>Preparation Assignment:</b> Must be completed before the exercises can be started. Most of these will include writing a simple C++ program. Please look at the end of each chapter in the text for the assignment. Once you have turned in the Preparation Assignment, you will be ready to work on the Exercise.</li>
<li><b>Questions &amp; Answers:</b> When class begins, we will spend about half an hour answering any questions you may have on the current chapter, previous chapter, the preparation assignment, or any other aspect of programming.</li>
<li><b>Problems:</b> The remainder of the class will be spent working through problems designed to extend your understanding of the topic. Some of these problems will be tricky, the common pitfalls new programmers encounter while programming in C++. Some will require a bit of creativity. Each problem is presented <g class="gr_ gr_121 gr-alert gr_gramm gr_inline_cards gr_run_anim Grammar multiReplace" id="121" data-gr-id="121">in</g> the end of the chapter.</li>
<li><b>Next Class</b>: The final 5 minutes of class will be spent giving you a peek at what we can expect for the next assignment and class period.</li>
</ol>
