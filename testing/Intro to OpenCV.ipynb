# Introduction to OpenCV


```python
# These are the libraries used in this notebook
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

## Displaying an Image

```python
img = cv2.imread('test.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title('Handsome Guy'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img = cv2.imread('test.jpg')
print ('RGB pixel values')
print (img)
```

```python
img = cv2.imread('test.jpg', 0)
print ('Grayscale image.  Each pixel is 0 to 255')
print (img)

plt.imshow(img, cmap='gray')
plt.title('Really Handsome Guy'), plt.xticks([]), plt.yticks([])
plt.show()
```

## Edge Detection Example

```python
img = cv2.imread('building.jpg',0)
edges = cv2.Canny(img,100,200)

plt.subplot(121), plt.imshow(img, cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])

plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```
